```python
#the import packages
import requests
import pandas as pd
from pandas import json_normalize
import requests
import os
from pathlib import Path
from datetime import datetime, timezone,timedelta,time
from scipy import stats
import json
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def loadDataFromFile(file_name):
    script_dir = Path().resolve().parent

    data_folder = script_dir / 'dataAnalysis and machine learning'/'data'
    print(data_folder)
    data_folder.mkdir(exist_ok=True)
    
    file_path = data_folder / (file_name + ".json")
    
    if file_path.exists():
        df = pd.read_json(file_path)
        print(f"Loaded {len(df)} records from {file_path}")
        return df
    else:
        print(f"File {file_path} does not exist.")
        return None    
```

```python
userInputDataRaw = loadDataFromFile("User:Calibration no source pollutant")
timeSeriesData = loadDataFromFile("Data:Calibration no source pollutant")
```

```python
userInputData = userInputDataRaw.drop(columns=["_id","timestamp","epoch_ms","userInputCategory","pollutant-type","quantity-used"])
```

```python
userInputData["timestamp_local"] = pd.to_datetime(userInputData["timestamp_local"])  # ensure datetime
userInputData["timestamp_local"] = userInputData["timestamp_local"].dt.floor("s")
```

```python
userInputData
```

```python
userInputDataCompressed = userInputData[userInputData["experimentState"].isin(["NoSourcePollutantInserted","InsertingSourcePollutant"])].copy()

userInputDataCompressed["timestamp StartingExperiment"] = userInputData["timestamp_local"].shift(1).loc[userInputDataCompressed.index]
userInputDataCompressed["timestamp EndingExperiment"]= userInputData["timestamp_local"].shift(-1).loc[userInputDataCompressed.index]
userInputDataCompressed = userInputDataCompressed.reset_index(drop = True)
userInputDataCompressed = userInputDataCompressed.rename(columns={"timestamp_local":"timestamp InsertingSource"})

userInputDataCompressed["z-score=3 for Id=0:BME680:breathVocEquivalent"] = np.nan
userInputDataCompressed["z-score=3 for Id=1:BME680:breathVocEquivalent"] = np.nan
userInputDataCompressed["z-score=3 for Id=2:BME680:breathVocEquivalent"] = np.nan
```

```python
def createTheTitleForTheFigure(experiment_number):
    date = userInputDataCompressed.iloc[experiment_number]["timestamp InsertingSource"].date()
    
    if ((experiment_number > 0 )):
        previous_date = userInputDataCompressed.iloc[experiment_number - 1]["timestamp InsertingSource"].date()
        if (date != previous_date):
            print("........................................................\n\n")
    
    title = "Experiment at row " + str(experiment_number)+ "At "+date.strftime("%Y-%m-%d")+ f" BME680:breathVocEquivalent.\n"
    if (userInputDataCompressed.iloc[experiment_number]["experimentState"] == "InsertingSourcePollutant"):
        title= title + "at the pollution source position to be "
        if (pd.isna(userInputDataCompressed.iloc[experiment_number]['front-wall'])==False):
           title =title + f"{userInputDataCompressed.iloc[experiment_number]['front-wall']} meters from the front wall, "
        if (pd.isna(userInputDataCompressed.iloc[experiment_number]['back-wall'])==False):
           title =title +f"{userInputDataCompressed.iloc[experiment_number]['back-wall']} meters from the back wall, "
        if (pd.isna(userInputDataCompressed.iloc[experiment_number]['side-right-wall'])==False):
           title =title+ f"{userInputDataCompressed.iloc[experiment_number]['side-right-wall']} meters from the side right wall, "
        if (pd.isna(userInputDataCompressed.iloc[experiment_number]['side-left-wall'])==False):
           title =title+ f"{userInputDataCompressed.iloc[experiment_number]['side-left-wall']} meters from the side left wall, "
    elif (userInputDataCompressed.iloc[experiment_number]["experimentState"] == "NoSourcePollutantInserted"):  
        title =title + "without source insertion.\n"
        
        
    
    title =title +"at the room:"+userInputDataCompressed.iloc[experiment_number]["room"]+"\n"   
    title=title +"Στοιχεία για το μενωμένο πείραμα:"
    title= title+"Αντικείμενο που χρησιμοποιείται¨"+row["item-used"]+", "
    if (pd.isna(row["are-doors-opened"]) == False):
        title = title + "Οι πόρτες είναι ανοιχτές, "
    if (pd.isna(row["are-fans-on"]) == False):
        title = title + "Οι ανεμιστήρες είναι ενεργοποιημένοι, "
    if (pd.isna(row["are-people-inside"]) == False):
        title = title + "Βρίσκονται άνθρωποι μέσα, "
    if (pd.isna(row["are-windows-opened"]) == False):
        title = title + "Τα παράθυρα είναι ανοιχτά, "
    if (pd.isna(row["notes"])== False):
        title= title + "Σημειώσεις:"+ row["notes"]

    return title

def printData(dict_of_timeseries,plot_old_data = True):
    for experiment_number in dict_of_timeseries:
        title = createTheTitleForTheFigure(experiment_number)
            
       
        plt.figure(figsize=(18, 8))
        if plot_old_data:
            y_column_data = "VOC"
        else:
            y_column_data = "cutted_VOC"
            
        sns.lineplot(data=dict_of_timeseries[experiment_number], x="timestamp", y=y_column_data, hue="sensors")
        plt.xticks(rotation=45)
        
        plt.axvline(x=userInputDataCompressed.iloc[experiment_number]["timestamp InsertingSource timedelta"], color="red", linestyle="--", linewidth=2, label="timestamp of the inserting source")
        plt.title(title)
    
        plt.show()
    
        print("\n")
```

```python
#find the value of z=3 for each dataframe
def find_value_VOC_withz_score_equals_3(voc_column):
    
    mean = voc_column.mean()
    std =  voc_column.std()
    value_VOC_with_z_score_equals_3 = 3 * std + mean
    return value_VOC_with_z_score_equals_3
```

```python
dict_of_timeseries = {}
for index, row in userInputDataCompressed.iterrows():
    #get datetime
    start = row["timestamp StartingExperiment"]
    turning_point = row["timestamp InsertingSource"]
    end = row["timestamp EndingExperiment"]
    #Grab data
    particular_experiment = timeSeriesData[(timeSeriesData["timestamp"] >= start) & (timeSeriesData["timestamp"] <= end)]
    particular_experiment = particular_experiment.sort_values(by = ["timestamp"],ignore_index = True)
    dict_of_timeseries[index] = particular_experiment

```

```python
for index, row in userInputDataCompressed.iterrows():
    particular_experiment =  dict_of_timeseries[index]

    #make the dataframe from multiple columns to have the data for each sensor to just one column
    particular_experiment = particular_experiment.melt(
        id_vars=["timestamp"], 
        value_vars=[
            "Id=0:BME680:breathVocEquivalent",
            "Id=1:BME680:breathVocEquivalent",
            "Id=2:BME680:breathVocEquivalent"
        ],
        var_name="sensors", 
        value_name="VOC"
    )
    particular_experiment = particular_experiment.dropna(ignore_index = True)

    particular_experiment = particular_experiment.sort_values(by = ["timestamp"],ignore_index = True)
    dict_of_timeseries[index] = particular_experiment
```

```python
for index, row in userInputDataCompressed.iterrows():
#Now, we only have timestamps based of whenever the sensor took the value, but we want to have times 
    # Get full range of seconds from start to end timestamp
    particular_experiment =  dict_of_timeseries[index]
    start = row["timestamp StartingExperiment"]
    turning_point = row["timestamp InsertingSource"]

    end = row["timestamp EndingExperiment"]
    full_time_range = pd.date_range(start, end, freq="S")
    
    all_sensors = particular_experiment["sensors"].unique()
    
    full_index = pd.MultiIndex.from_product([full_time_range, all_sensors], names=["timestamp", "sensors"])
    extended_particular_experiment= pd.DataFrame(index=full_index).reset_index()
    particular_experiment = extended_particular_experiment.merge(particular_experiment, on=["timestamp", "sensors"], how="left",suffixes=["",None])
    particular_experiment["after_insertion"] = particular_experiment["timestamp"] > turning_point
    particular_experiment["original_value"] = pd.notna(particular_experiment["VOC"])
    dict_of_timeseries[index] = particular_experiment
```

```python
userInputDataCompressed["actual timestamp StartingExperiment"] = np.nan
userInputDataCompressed["actual timestamp EndingExperiment"] = np.nan

for index, row in userInputDataCompressed.iterrows():
    particular_experiment =  dict_of_timeseries[index]
    #print(f"index:{index}")
    first_timestamps =  particular_experiment[particular_experiment["VOC"].notna()].groupby("sensors", as_index=False)["timestamp"].min()
   # print(f"first_timestamps:{first_timestamps}")
    
    last_timestamps =  particular_experiment[particular_experiment["VOC"].notna()].groupby("sensors", as_index=False)["timestamp"].max()
    #print(f"last_timestamps:{last_timestamps}")

    userInputDataCompressed.at[index,"actual timestamp StartingExperiment"] = first_timestamps["timestamp"].max()
    
    userInputDataCompressed.at[index,"actual timestamp EndingExperiment"] = last_timestamps["timestamp"].min()

   

    start = userInputDataCompressed.at[index, "actual timestamp StartingExperiment"]
    end = userInputDataCompressed.at[index, "actual timestamp EndingExperiment"]
 
            
    particular_experiment =   particular_experiment.loc[particular_experiment["timestamp"].between(start, end,inclusive ="both")]
        
    particular_experiment =particular_experiment.reset_index(drop=True)
   
    dict_of_timeseries[index] = particular_experiment
```

```python
for index, row in userInputDataCompressed.iterrows():
    
        particular_experiment =  dict_of_timeseries[index]

        particular_experiment["cutted_VOC"] = particular_experiment["VOC"]
        particular_experiment["zscore"] =np.nan
        sensors = particular_experiment["sensors"].unique()
        for sensor in all_sensors:
            inner_mask = particular_experiment["sensors"] == sensor
            
            outer_mask = particular_experiment.loc[inner_mask, "VOC"] 
            sensor_full_name = "z-score=3 for "+sensor
            particular_experiment.loc[inner_mask,"zscore"] = stats.zscore(outer_mask)
            userInputDataCompressed.at[index,sensor_full_name] = find_value_VOC_withz_score_equals_3(outer_mask)
            
            
            particular_experiment.loc[ 
                inner_mask & (particular_experiment["zscore"] > 3)
                ,"cutted_VOC"] =userInputDataCompressed.at[index,sensor_full_name] 
            
            
        dict_of_timeseries[index] = particular_experiment
```

```python
userInputDataCompressed["timestamp InsertingSource timedelta"] = np.nan
for index, row in userInputDataCompressed.iterrows():

    particular_experiment =  dict_of_timeseries[index]
    sensors = particular_experiment["sensors"].unique()
    for sensor in sensors:
        
        mask = particular_experiment["sensors"]==sensor
        masker = particular_experiment.loc[mask]
        masker_condition = masker.loc[masker["after_insertion"] == True]
        first_index = masker_condition.index[0]
        insertion_time = masker.at[first_index,"timestamp"]
        
        userInputDataCompressed.at[index,"timestamp InsertingExperiment timedelta"] = insertion_time
    dict_of_timeseries[index] =   particular_experiment
    
```
